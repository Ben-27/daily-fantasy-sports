```python
from Game import *
import pandas as pd

%load_ext autoreload
%autoreload 2
```

```python
week = Game('FanDuel-NFL-2022 ET-11 ET-13 ET-82959-players-list.csv',
            'NFL',
            '2022-10 Fantasy Pros Projections 22-11-12.xlsx',
            'FantasyPros NFL Historical.xlsx',
            2022, 10)
```

```python
DROP = ['Minnesota Vikings', 'Los Angeles Chargers', 'Green Bay Packers', 'Miami Dolphins', # NFL
        'Seattle Seahawks', 'Cleveland Browns', 'Jacksonville Jaguars'] # SI

week.generateRosters('reimagined', drop_players_initial=DROP)
```

```python
# verify dropped players
week.tracking['preprocessing_drops'].sort_values('Projected', ascending=False).head(10)
```

```python
top = week.selectRosters(.75)
week._selection
```

```python
# players with appearances
x = week.data[['Nickname', 'Position', 'Salary', 'Projected', 'Top Appearances']]
x[x['Top Appearances'] > 0].sort_values('Top Appearances', ascending=False)
```

```python
# highest scoring rosters and frequency
y = top.copy()
y['Roster'] = y.index
y_g = y.groupby('Score', as_index=False).agg({'Roster':'first', 'Salary':'first', 'Projected':'first', 'STD':'first', 'K':min, 'Score':['first', 'count']})
y_g.sort_values(('Score', 'first'), ascending=False).head(10)
```
